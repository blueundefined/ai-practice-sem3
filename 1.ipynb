### –ó–∞–¥–∞–Ω–∏–µ 4.3.2
–î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è. –°–æ–∑–¥–∞—Ç—å –º–∞—Å—Å–∏–≤ –∏–∑ 10 –∑–Ω–∞—á–µ–Ω–∏–π —Ñ—É–Ω–∫—Ü–∏–∏ (ùë•, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –æ—Ç 1 –¥–æ 10). –í—ã–¥–µ–ª–∏—Ç—å —Å—Ä–µ–∑ –ø–µ—Ä–≤–æ–π –ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ ‚Äì –ª–∏–Ω–µ–π–Ω—ã–π –∏ –¥–ª—è —Å—Ä–µ–∑–∞ ‚Äì —Ç–æ—á–µ—á–Ω—ã–π.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sqrt(1 + np.e**np.sqrt(x) + np.cos(x**2)) / np.abs(1 - np.sin(x)**3) + np.log(np.abs(2*x))

x = np.linspace(1, 10, 10)
y = f(x)

y_slice = y[:5]

plt.plot(x, y, 'b-', label='–ú–∞—Å—Å–∏–≤')

plt.plot(x[:5], y_slice, 'r o', label='–°—Ä–µ–∑')

plt.legend(loc="best")
plt.show()
```

### –ó–∞–¥–∞–Ω–∏–µ 4.3.3
–î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è |cos(xe(cos(x)+ln(x+1)))| 
–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫ –Ω–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ (0,10) —Å —à–∞–≥–æ–º 1 —Å –∑–∞–ª–∏–≤–∫–æ–π –ø–ª–æ—â–∞–¥–∏ –∏ –Ω–∞–π—Ç–∏ —ç—Ç—É –ø–ª–æ—â–∞–¥—å –ø–æ–¥ –Ω–∏–º. 
–î–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –ø–ª–æ—â–∞–¥–∏ –≤–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏–∏ trapz(), –≤—ã—á–∏—Å–ª—è—é—â–µ–π –∏–Ω—Ç–µ–≥—Ä–∞–ª –ø–æ –ø—Ä–∞–≤–∏–ª—É —Ç—Ä–∞–ø–µ—Ü–∏–∏.
–î–ª—è –µ–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–π —Ä–∞–±–æ—Ç—ã –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø–æ–¥–∫–ª—é—á–∏—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏:
from scipy.integrate import simps
from numpy import trapz

```python
!pip install scipy
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
from numpy import trapz

def f(x):
    return np.abs(np.cos(x * np.exp(np.cos(x) + np.log(x + 1))))
 
x = np.arange(0, 10, 1)  
y = f(x)

plt.plot(x, y, 'g-', label="–§—É–Ω–∫—Ü–∏—è")
plt.fill_between(x, 0, y, where=(y >= 0), color='lightgreen', alpha=0.5)

area = trapz(y, x)

print("–ü–ª–æ—â–∞–¥—å –ø–æ–¥ –≥—Ä–∞—Ñ–∏–∫–æ–º:", area)

plt.legend(loc='best')
plt.show()
```

### –ó–∞–¥–∞–Ω–∏–µ 4.3.5
–°–æ–∑–¥–∞–π—Ç–µ –ø—Ä–æ—Å—Ç–µ–π—à–∏–π –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä, –≤–∫–ª—é—á–∞—é—â–∏–π –æ—Å–Ω–æ–≤–Ω—ã–µ –¥–µ–π—Å—Ç–≤–∏—è –¥–ª—è –¥–≤—É—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö ‚Ä≤+‚Ä≤, ‚Ä≤‚àí‚Ä≤, ‚Ä≤*‚Ä≤, ‚Ä≤√∑‚Ä≤, –∞ —Ç–∞–∫–∂–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å–ª–µ–¥—É—é—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π: exp(x+y), sin(x+y), cos(x+y), x^y.

```python
import numpy as np

print("–î–æ—Å—Ç—É–ø–Ω—ã–µ –æ–ø–µ—Ä–∞—Ü–∏–∏:")
print("1. –°–ª–æ–∂–µ–Ω–∏–µ")
print("2. –í—ã—á–∏—Ç–∞–Ω–∏–µ")
print("3. –£–º–Ω–æ–∂–µ–Ω–∏–µ")
print("4. –î–µ–ª–µ–Ω–∏–µ")
print("5. e^(x+y)")
print("6. sin(x+y)")
print("7. cos(x+y)")
print("8. x^y")
print("9. –ó–∞–∫—Ä—ã—Ç—å")
print("\n")

while True:
    choice = input("–í—ã–±–µ—Ä–∏—Ç–µ –æ–ø–µ—Ä–∞—Ü–∏—é: ")

    if choice == '9':
        break

    if choice in ('1', '2', '3', '4', '5', '6', '7', '8'):
        x = float(input("–í–≤–µ–¥–∏—Ç–µ –ø–µ—Ä–≤–æ–µ —á–∏—Å–ª–æ: "))
        y = float(input("–í–≤–µ–¥–∏—Ç–µ –≤—Ç–æ—Ä–æ–µ —á–∏—Å–ª–æ: "))

        if choice == '1':
            result = np.add(x, y)
        elif choice == '2':
            result = np.subtract(x, y)
        elif choice == '3':
            result = np.multiply(x, y)
        elif choice == '4':
            if y == 0:
                result = "–û—à–∏–±–∫–∞: –Ω–∞ –Ω–æ–ª—å –Ω–µ –¥–µ–ª–∏–º"
            else:
                result = np.divide(x, y)
        elif choice == '5':
            result = np.exp(x + y)
        elif choice == '6':
            result = np.sin(x + y)
        elif choice == '7':
            result = np.cos(x + y)
        elif choice == '8':
            result = np.power(x, y)
        print(f"–û–ø—Ü–∏—è {choice}, x = {x}, y = {y}")
        print("–†–µ–∑—É–ª—å—Ç–∞—Ç:", result, "\n")
    else:
        print("–ù–µ–≤–µ—Ä–Ω–∞—è –æ–ø–µ—Ä–∞—Ü–∏—è.")
```
