### Задание 1.3.1

Дан массив данных
```python
X = np.array([
    [5,3],
    [10,15],
    [15,12],
    [24,10],
    [30,45],
    [85,70],
    [71,80],
    [60,78],
    [55,52],
    [80,91],
    ])
```
Требуется провести кластерный анализ данных методом k-средних. 
Поэкспериментируйте с количеством кластеров.

````python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X = np.array([[5,3],[10,15],[15,12],[24,10],[30,45],[85,70],[71,80],[60,78],[55,52],[80,91],])

num_clusters = [2, 3, 4, 5, 6]

for n_clusters in num_clusters:
    kmeans = KMeans(n_clusters=n_clusters, n_init='auto', random_state=0).fit(X)
    labels = kmeans.labels_
    plt.figure()
    plt.scatter(X[:, 0], X[:, 1], c=labels)
    plt.title('Количество кластеров: %d' % n_clusters)
    plt.show()
````

### Задание 1.3.2 
Выполните кластеризацию для набора данных ирисов Фишера. 
Выполните предсказания для модели. Поэкспериментируйте с количеством кластеров.

````python
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Загружаем набор данных ирисов Фишера
iris = load_iris()
X = iris.data

# Определяем количество кластеров
num_clusters = [2, 3, 4, 5, 6]

# Проходим по каждому количеству кластеров
for n_clusters in num_clusters:
    kmeans = KMeans(n_clusters=n_clusters, n_init='auto', random_state=0).fit(X)
    labels = kmeans.labels_
    plt.figure()
    plt.scatter(X[:, 0], X[:, 1], c=labels)
    plt.title('Количество кластеров: %d' % n_clusters)
    plt.show()
````

### Задание 2.3
 Выполните иерархическую кластеризацию для набора данных об ирисах Фишера. При этом необходимо использовать любые два признака (всего их четыре).

````python
from sklearn import datasets
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
from sklearn.metrics import adjusted_rand_score
import matplotlib.pyplot as plt

iris = datasets.load_iris()
X = iris.data[:, :2] 

cluster = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')

cluster_labels = cluster.fit_predict(X)

silhouette_avg = silhouette_score(X, cluster_labels)
print("Коэффициент силуэта:", silhouette_avg)

adjusted_rand_score = adjusted_rand_score(iris.target, cluster_labels)
print("Скорректированный балл:", adjusted_rand_score)

plt.scatter(X[:,0],X[:,1], c=cluster_labels, cmap='rainbow')
plt.show()
````
